```python
from ultralytics import YOLO
import yaml, os

# YAML 파일 로드
yaml_path = "day_data.yaml"
with open(yaml_path, encoding="utf-8") as f:
    data = yaml.safe_load(f)

root = data.get("path", "")
train_path = os.path.join(root, data["train"])
val_path = os.path.join(root, data["val"])

print(f"Train path: {train_path}")
print(f"Val path:   {val_path}")

# 폴더 및 파일 존재 여부 확인
print("\n=== 경로 검증 ===")
for p in [train_path, val_path]:
    if not os.path.exists(p):
        print(f"존재하지 않음: {p}")
    else:
        print(f"경로 확인됨: {p}")
        # 예시로 몇 개 파일만 출력
        for r, _, files in os.walk(p):
            for f in files[:3]:
                print("   └", os.path.join(r, f))
            break
```

```python
import json, os

json_path = "./dataset/traffic_data/training/bbox_tr/annotations/day_train.json"
base_path = "./dataset/traffic_data"

with open(json_path, encoding="utf-8") as f:
    coco = json.load(f)

missing = []
for i, img in enumerate(coco["images"][:10]):  # 처음 10개만 테스트
    file_path = os.path.join(base_path, img["file_name"])
    if not os.path.exists(file_path):
        missing.append(file_path)

if missing:
    print("존재하지 않는 이미지 경로 예시:")
    print("\n".join(missing[:5]))
else:
    print("모든 샘플 이미지 경로가 정상적으로 존재합니다.")
```
