```python
from ultralytics import YOLO
from typing import Any
from pathlib import Path

# Îç∞Ïù¥ÌÑ∞ÏÖã YAML Í≤ΩÎ°ú
OUT_ROOT: Path = Path("yolo_transfer_all") # ÌïÑÏöîÏãú Í≤ΩÎ°ú Î≥ÄÍ≤Ω!!
data_yaml: Path = OUT_ROOT / "all_data" / "data.yaml"
```

```python
# Ï†ÑÏù¥ÌïôÏäµ
model: YOLO = YOLO("yolo11m.pt")
results: Any = model.train(
    data=str(data_yaml),
    epochs=40,
    batch=6,                   
    imgsz=608,
    rect=True,
    optimizer="AdamW",
    lr0=7e-4,
    cos_lr=True,
    amp=True,
    workers=4,
    device="cuda",
    seed=0,
    freeze=10,
    patience=30,
    # Ï¶ùÍ∞ï Í¥ÄÎ†®
    # mosaic=0.0,
    # mixup=0.0,
    project="runs_all/all_stage",
    name="exp_day"
)
# 31 epoch 
# Ï†ÑÏù¥ÌïôÏäµ Í≤∞Í≥º Í≤ΩÎ°ú
result_model: str = "runs_all/all_stage/exp_day/weights/last.pt"
print(f"model last -> {result_model}")
```
