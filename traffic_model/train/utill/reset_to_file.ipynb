```python
import os
import subprocess

trash_path = os.path.expanduser("~/.local/share/Trash")

if os.path.exists(trash_path):
    trash_size = subprocess.getoutput(f"du -sh {trash_path}")

    file_count = sum(len(files) for _, _, files in os.walk(os.path.join(trash_path, "files")))
    
    print(f"휴지통 경로: {trash_path}")
    print(f"총 용량: {trash_size}")
    print(f"파일 개수: {file_count}")
else:
    print("휴지통 폴더가 없습니다.")



###2
import os
import shutil

trash_path = os.path.expanduser("~/.local/share/Trash")

if os.path.exists(trash_path):
    try:
        shutil.rmtree(trash_path)
        os.makedirs(os.path.join(trash_path, "files"), exist_ok=True)
        os.makedirs(os.path.join(trash_path, "info"), exist_ok=True)
        print("휴지통이 성공적으로 비워졌습니다.")
    except Exception as e:
        print("오류 발생:", e)
else:
    print("휴지통 폴더가 존재하지 않습니다.")
```
